# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Creat basic swarmplot

```python
sns.swarmplot(sales, y='Total')
```

# 4. Create basic swarmplot and GROUP it by payment

```python
plt.figure(figsize=(10,5))
sns.swarmplot(sales, x='Payment', y='Total', hue='Payment').tick_params('x', rotation=45)
```

# 5. Create swarmplot and GROUP it with HUES

```python
plt.figure(figsize=(10,5))
sns.swarmplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], dodge=True)
```

# 6. Customizing markers on swarmplot

```python
plt.figure(figsize=(10, 10))
sns.swarmplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], palette='dark', marker='*', size=8).grid(True)
```

# 7. Create swarmplot on top of boxplot

```python
sns.boxplot(sales, x='Payment', y='Total', color='grey')
sns.swarmplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], palette='dark', alpha=.8)
```

```python
sns.violinplot(sales, x='Payment', y='Total', color='grey')
sns.swarmplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], palette='Paired', alpha=.8)
```
