# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Creating basic catplot

```python
sns.catplot(sales, x='Payment', y='Total') # stripplot as default
```

# 4. Customizing which graph will be shown

```python
sns.catplot(sales, x='Customer type', y='Total', kind='swarm')
```

# 5. Catplot using col parameter

```python
 sns.catplot(sales, x='Payment', y='Total', col='Gender')
```

Bakal ada facet graph yang misahin mereka berdasarkan gender, bentuknya kolom

# 6. Catplot using col and row parameter

```python
sns.catplot(sales, x='Payment', y='Total', col='Gender', row='City')
```
