# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Make basic violinplot

```python
sns.violinplot(sales, y='Total')
```

# 4. Create a violinplot with 2 categorical and 1 numeric variable

```python
sns.violinplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], split=True)
```

# 5. Change the box in violinplot to Horizontal line

```python
sns.violinplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], split=True,
               inner='quart')
```

# 6. Draw line for each observations in violinplot

```python
sns.violinplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], split=True,
               inner='stick')
```

# 7. Change the amount of smoothing the using bw attribute

```python
sns.violinplot(sales, x='Payment', y='Total', hue='Gender', split=True, inner='quart', bw_method=.2)
```

# 8. Cut the extreme value

```python
sns.violinplot(sales, x='Payment', y='Total', hue='Gender', hue_order=['Male', 'Female'], split=True, inner='quart', cut=-2)
```
