# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Making basic boxplot in NUMERIC variable

```python
sns.boxplot(sales, y='Total', width=0.3)
```

# 4. Makin basic boxplot in CATEGORICAL variable

```python
sns.boxplot(sales, x='Payment', y='Total', hue='Payment', width=0.3)
```

# 5. Making basic boxplot in 2 CATEGORICAL with HUE

```python
sns.boxplot(sales, x='Payment', y='Total', hue='Gender', width=0.3,
            hue_order=['Male', 'Female'])
```

# 6. Adding mean marker and style it using MEANPROPS

```python
sns.boxplot(sales, x='Payment', y='Total', hue='Gender', width=0.3, 
            showmeans=True, meanprops={'marker':'o', 'markerfacecolor':'white', 'markeredgecolor':'blue'},
            hue_order=['Male', 'Female'])
```

# 7. Make HORIZONTAL boxplot

```python
sns.boxplot(sales, x='Total', y='Payment', hue='Gender',
            hue_order=['Male', 'Female'], showmeans=True,
            meanprops={'markeredgecolor':'white',
                       'marker':'|', 'markersize':'30'})
```

# 8. Palette and linewidth

```python
sns.boxplot(sales, x='Total', y='Payment', hue='Gender',
            hue_order=['Male', 'Female'], showmeans=True,
            meanprops={'markeredgecolor':'white',
                       'marker':'|', 'markersize':'30'},
            palette='Purples', linewidth=1.3)
```

# 9. Making boxplot for EACH numeric variable in dataframe

```python
plt.figure(figsize=(10,5))
sns.boxplot(sales)
```
