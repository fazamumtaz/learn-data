# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Make a countplot

## 1. Vertikal countplot

```python
plt.figure(figsize=(15,5))
ax = sns.countplot(sales, x='Product line', hue='Product line', palette='pastel')
ax.tick_params(axis='x', rotation=45)
ax.set_xlabel('Product Category')
plt.show()
```

## 2. Horizontal countplot

```python
sns.countplot(sales, y='Product line', hue='Product line', palette='tab10')
```

## 3. Using hue to categorize another variabel

```python
sns.countplot(sales, x='Product line', hue='Gender').tick_params(axis='x', rotation=45)
```

## 4. Using palette for the color

```python
sns.countplot(sales, x='Product line', hue='Gender', palette='Pastel2').tick_params('x', rotation=45)
```

## 5. Styling using edgecolor, linewidth and facecolor

```python
sns.countplot(sales, x='Product line',
              facecolor=(0, 0, 0, 0),
              linewidth=5, 
              edgecolor=sns.color_palette('dark', 3)).tick_params('x', rotation=45)
```
