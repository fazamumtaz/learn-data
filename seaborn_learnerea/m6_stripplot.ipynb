# 1. Setup Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. Load Dataset

```python
sales = pd.read_excel('../dataset/supermarket_sales.xlsx')
sales.head()
```

# 3. Creating basic stripplot

```python
sns.stripplot(sales, x='Payment', y='Total', color='green')
```

# 4. Expanding markers using jitter

```python
sns.stripplot(sales, x='Payment', y='Total', color='green', alpha=.3, jitter=.2)
```

# 5. Draw line around the markers using linewidth

```python
sns.stripplot(sales, x='Payment', y='Total', alpha=.3, linewidth=1)
```

# 6. Include third categorical variable using HUE and DODGE

```python
sns.stripplot(sales, x='Payment', y='Total', hue='Gender', dodge=True, jitter=.23, alpha=.5)
```

# 7. Draw a stripplot on top of Violinplot

```python
sns.stripplot(sales, x='Payment', y='Total', jitter=.15, alpha=.6, linewidth=.5, color='pink')
sns.violinplot(sales, x='Payment', y='Total', color='lightblue')
```
