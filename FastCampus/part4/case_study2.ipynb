```python
import pandas as pd

# DataFrame order_history
order_history = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5, 6, 7, 8],
    "cust_no": ["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8"],
    "order_amount": [100, 200, 250, 150, 300, 350, 100, 150]
})

# DataFrame customer_base
customer_base = pd.DataFrame({
    "cust_id": ["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9", "C10"],
    "city": ["Bandung", "Jakarta", "Tangerang", "Tangerang", "Jakarta", "Jakarta", "Jakarta", "Bandung", "Tangerang", "Jakarta"]
})
```

```python
# Merging those 2 dfs
data = customer_base.merge(order_history, how='inner', left_on='cust_id', right_on='cust_no').reset_index(drop=True)
data
```

```python
# finding the total order by city
data.groupby('city').agg({'order_amount':'sum'}).sort_values(by='order_amount', ascending=False)
```
