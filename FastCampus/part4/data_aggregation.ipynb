# Data Aggregation

## Value Counts

```python
import pandas as pd

iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris.info()
```

```python
# Value Counts
iris['species'].value_counts()
```

```python
# Groupby
iris.groupby(['species']).count()
```

```python
iris.groupby(['species']).min()
```

```python
iris.groupby(['species']).max()
```

```python
iris.groupby('species').sum()
```

```python
iris.groupby('species').mean()
```

```python
iris.groupby('species').median()
```

```python
iris.groupby('species').describe().T #T for transpose? haha
```

```python
# Akan menjadi masalah saat groupby, contoh di kolom spesies akna menjadi multilevel column, jadi nanti gabisa dipake buat filtering with criteria
df = iris.groupby('species').mean().reset_index()
df
```

```python
df[df['species'] == 'setosa']
```

### groupby() with 2 or more aggregations

```python
iris.groupby('species').agg(['min', 'max', 'mean'])
```

```python
iris.groupby('species').agg({'sepal_length':'mean', 'sepal_width':['min', 'max']})
```
