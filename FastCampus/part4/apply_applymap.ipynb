## Apply

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': [20, 40, 60, 80]
})

df
```

```python
def sum(x):
    result = x.sum()
    return result
df['d'] = df.apply(sum, axis=1)
df
```

```python
df.loc[4, :] = df.apply(sum, axis=0) #adding new row
df
```

```python
def multipleby2(x):
    return x*2
df['e'] = df['B'].apply(multipleby2)
df
```

```python
data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': [20, 40, 60, 80]
})

data
```

```python
# using lambda
```

```python
data['d'] = data.apply(lambda x:x.sum(), axis=1)
data
```

```python
data.loc[4,:] = data.apply(lambda x:x.sum(), axis=0)
data
```

```python
data['E'] = data['C'].apply(lambda x:x*2)
data
```
