## Continous Plots

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

### Histogram

Argumen yang bisa diterima `sns.histplot()`:

* **bins**: Untuk menentukan jumlah bin (interval).
* **binwidth**: Menentukan lebar bin.
* **kde**: Menambahkan kurva KDE (Kernel Density Estimation).
* **hue**: Membagi data berdasarkan kategori.
* **multiple**: Cara menampilkan hue.
* **element**: Mengontrol tampilan bar (bars(default), step(garis tepi saja))
* **stat**: Mengubah skala histogram.

```python
sns.histplot(penguins, x='body_mass_g', bins=20, binwidth=200, kde=True, color='#48A6A7')
```

```python
sns.histplot(penguins, x='body_mass_g', hue='species', multiple='stack', element='bars', fill=True, palette='Set2')
```

**Penjelasan:**
 + **hue:** mengelompokkan berdasarkan species penguin
 + **multiple:**
   + **layer:**  saling menutupi satu sama lain (berisiko tidak terlihat)
   + **stack:** bertingkat
   + **dodge:** terpisah setiap kategori
+ **element:** bar(default), step(outline setiap graph)
+ **fill:** True(default), False(graph with outline only)

```python
sns.histplot(penguins, x='body_mass_g', stat='frequency', shrink=0.8)
```

**Penjelasan:**
+ **"count"**  Jumlah data (default).
+ **"density"**  Normalisasi sehingga total luasnya = 1.
+ **"probability"**  Setiap bar mewakili probabilitasnya.

### Boxplot (box-and-whisker-plot):

Boxplot menggambarkan ringkasan statistik 5 angka (Five Number Summary) dalam satu grafik:

+ **Min (minimum)**  Nilai terkecil (batas bawah tanpa outlier)
+ **Q1 (Kuartil 1)**  Nilai 25% data terendah
+ **Median (Q2)**  Nilai tengah data
+ **Q3 (Kuartil 3)**  Nilai 75% data terendah
+ **Max (maximum)**  Nilai terbesar (batas atas tanpa outlier)
+ **Outlier**  Ditampilkan sebagai titik di luar whisker jika data terlalu jauh dari Q1 atau Q3.

```python
sns.boxplot(data=penguins, x='body_mass_g', width=0.2, color='#48A6A7')
```

```python
sns.boxplot(data=penguins, y='species', x='body_mass_g', width=0.2, color='#48A6A7', fliersize=2, showfliers=False)
```
