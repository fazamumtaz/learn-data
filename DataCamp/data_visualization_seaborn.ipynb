### Adding a third variable with hue

we're gonna use the tip dataset from sns' dataset

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill',
                y='tip',
                data=tips)
```

normally, the normal scatterplot will show, and If we want to see which one is smoker which not, we can use the hue argument

```python
sns.scatterplot(x='total_bill',
                y='tip',
                hue='smoker',
                data=tips)
```

### Hue order with hue_order argument

```python
sns.scatterplot(x='total_bill',
                y='tip',
                hue='smoker',
                hue_order=['No', 'Yes'],
                data=tips)
```

### Customizing color with palette()

```python
colors = {
    'Yes':'blue',
    'No':'red'
}
sns.scatterplot(x='total_bill',
                y='tip',
                hue='smoker',
                hue_order=['Yes', 'No'],
                palette=colors,
                data=tips)
```

```python
sns.countplot(x='smoker', data=tips, hue='sex')
```
