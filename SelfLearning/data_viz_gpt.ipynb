# Beginner Level: Basic Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sales = pd.read_csv('../dataset/sales_data.csv')
sales.head()
```

## Line Plot: Total Revenue

### Using Figure and Axes

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Grouping revenue per bulan
monthly_sales = sales.groupby('Purchase_Month')['Revenue'].sum().reindex(month_order)

fig, ax = plt.subplots()
ax.plot(monthly_sales, marker='o', color='r', label='Revenue 2024')
ax.set_title('Total Revenue per Month', 
             fontsize=16, 
             fontweight='bold', 
             color='darkblue')
ax.set_xlabel('Month')
ax.set_ylabel('Revenue (USD)')
ax.tick_params(axis='x', rotation=90)
ax.grid(True, linestyle='--', alpha=0.3)
ax.legend(loc='upper left')
# ax.annotate('Max Revenue', xy=(8, 230000), xytext=(10, 23000))
ax.annotate('Max Revenue', xy=(7.3, 231000), xytext=(8.5, 228000), arrowprops={'arrowstyle':'->', 'color':'blue'})
```

### Using plt.plot()

```python
# Plot the graph
plt.plot(monthly_sales, color='red', linestyle='--', marker='o', markersize=6, label="Revenue 2024")

# Customizing Label
plt.title('Total Revenue per Month', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Month', fontsize=12, color='gray')
plt.ylabel('Revenue (USD)', fontsize=12, color='gray')

# Other customization
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend(loc='upper left')

plt.annotate('Titik Maksimum', xy=(5, 12000), xytext=(3, 15000), 
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```

## Bar Chart: Total Revenue by Category

### Using plt.bar()

```python
revenue_cat = sales.groupby('Category')['Revenue'].sum()
# revenue_cat.index
bar_labels = ['red', 'blue', '_red', 'orange', 'yellow']

plt.bar(revenue_cat.index, revenue_cat.values, color='orange', alpha=0.5, color=bar_labels)
plt.show()
```
